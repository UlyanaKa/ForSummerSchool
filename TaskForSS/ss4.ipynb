```python
import re
import pandas as pd
```

Текста были получены утомительным парсингом

```python
flnms = ['Ёж.txt', 'Готы.txt', 'Язык.txt']
changes = pd.read_csv('changes.csv')
changes['from'][18] = 'ёж' # bcs ёж!=еж   
```

```python
for flnm in flnms:
  with open(flnm) as fl: # open every file
    text = fl.read()
    for i in range(len(changes['from'])): # change every word from list
      frm = changes['from'][i]
      to = changes['to'][i]
      
      text = re.sub(f'\\b{frm}\\b', to, text)
      text = re.sub(f'\\b{frm.capitalize()}\\b', to.capitalize(), text) #if it's capitalized, change to to.capitalize()
  with open(flnm[:-4]+'_edited.txt', 'w') as fl: # writing edit in file_edited
    fl.write(text)   
```
